# Complex Patterns & Sequences

### Print a hollow diamond pattern with numbers where outer edge has sequential numbers and inner hollow shows calculation n × row.

```python


n = 5

rows = list(range(1, n+1)) + list(range(n-1, 0, -1))

for row in rows:
    spaces = " " * (n - row)
    if row == 1:
        print(spaces + str(n * row))
    else:
        inner = " " * (2 * row - 3)
        print(spaces + str(row) + inner + str(n * row))
```

### Generate Floyd’s triangle but with prime numbers only. Each row should contain the next onsecutive primes.

```python
n = 5

def is_prime(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def prime_collection(num):
    primes = []
    for i in range(2, num + 1):
        if is_prime(i):
            primes.append(i)
    return primes

primes = prime_collection(n)
for i in range(1, len(primes)):
    for j in range(i):        
        print(primes[j], end=" ")
    print()


```
