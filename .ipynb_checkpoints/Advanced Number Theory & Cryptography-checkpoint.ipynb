### Implement the Sieve of Eratosthenes without using arrays - use nested loops to mark and identify primes up to n=10000. Count twin primes (primes differing by 2).

```python
def sieveEratosthenes(num):
    prime_numbers={i: True for i in range(num)}
    root_num=2
    
    while root_num*root_num<=num:
        if prime_numbers[root_num]:
            for i in range(root_num * root_num, num + 1,root_num):
                prime_numbers[i] = False
            root_num += 1
        root_num=root_num+1

    res = []
    for root_num in range(2, num + 1):
        if prime_numbers[root_num]:
            res.append(root_num)
    
    return res
        
    

sieveEratosthenes(35)
```

### Implement the Chinese Remainder Theorem for two congruences using only loops and conditionals. Solve: x ≡ a1 (mod m1 ) and x ≡ a2 (mod m2 )`

```python
def chinese_remainder_theorem(a1, m1, a2, m2):
    
    x = 0
    while True:
        if x % m1 == a1 % m1 and x % m2 == a2 % m2:
            return x
        x += 1


chinese_remainder_theorem (2, 3, 3, 5)

```

### Implement Fermat’s primality test: check if ap−1 ≡ 1 (mod p) for multiple random values of a.Identify Carmichael numbers (pseudoprimes).

```python
import random

def fermat_test(n, k=5):
    if n < 2: return False
    for _ in range(k):
        a = random.randint(2, n - 1)
        if pow(a, n - 1, n) != 1:
            return False
    return True

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

def is_carmichael(n):
    if n < 2 or is_prime(n): return False
    temp, d, factors = n, 2, []
    while d * d <= temp:
        if temp % d == 0:
            factors.append(d)
            temp //= d
            if temp % d == 0: return False 
        d += 1
    if temp > 1: factors.append(temp)
    return all((n - 1) % (p - 1) == 0 for p in factors)


print("Fermat Primality \n")
for n in [17, 100, 561, 1729]:
    print(f"  n={n:>5}  fermat={fermat_test(n)}  actually_prime={is_prime(n)}  carmichael={is_carmichael(n)}")

carmichaels = [n for n in range(2, 10001) if is_carmichael(n)]
print(f"\nCarmichael numbers up to 10,000:\n  {carmichaels}")
```
