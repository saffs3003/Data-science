```python
def binary_exp(base, expo, mod):
    bits=bin(expo)[2:][::-1]
    
    res=1
    base %= mod
    for bit in bits:
        if bit=='1':
            res=(res*base)%mod
        base=(base*base)%mod
            
    return res


            
            

        


binary_exp(3, 1000000, 1000000007)
```

### Implement a program that checks if a given number is a perfect square without using square root function. Use binary search approach with while loop.

```python
def perfect_square(num):
    left=1
    right=num
    while left<=right:
        mid = int(left + (right - left) / 2)
        square=mid*mid
        if square==num:
            return True
        elif square<num:
            left=mid+1
        elif square>num:
            right=mid-1
    return False

perfect_square(64)
```
